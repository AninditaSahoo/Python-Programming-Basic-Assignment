<b>Q1. </b>Write a Python program to check if the given number is a Disarium Number?

```python
def CheckDisarium(num):
    res = 0
    for i in range(0,len(num)):
        res = res + int(num[i])**(i+1)
    if res == int(num):
        return num+" is Disarium"
    else:
        return num+" is Not Disarium"
num = input("Enter a number : ")
CheckDisarium(num)
```

<b>Q2. </b>Write a Python program to print all disarium numbers between 1 to 100?

```python
l = [] 
for j in range(1,100):
    num = str(j)
    res = 0
    for i in range(0,len(num)):
        res = res + int(num[i])**(i+1)
    if res == int(num):
        l.append(j)
print("The disarium numbers between 1 to 100 are : {}".format(l))
```

<b>Q3. </b>Write a Python program to check if the given number is Happy Number?

```python
def sum_of_square(digit):
    num = str(digit)
    res = 0
    for i in num:
        res = res+int(i)**2 
    return res
num = int(input("Enter a number : "))
result = num;     
while(result != 1 and result != 4):    
    result = sum_of_square(result)
if result == 1:
    print(str(num) + " is a happy number")
elif(result == 4):
    print(str(num) + " is not a happy number");    
```

<b>Q4. </b>Write a Python program to print all happy numbers between 1 and 100?

```python
def sum_of_square(digit):
    num = str(digit)
    res = 0
    for i in num:
        res = res+int(i)**2 
    return res
ans = []
for i in range(1,100):
    result = i
    while(result != 1 and result != 4):    
        result = sum_of_square(result)
    if result == 1:
        ans.append(i)
print("The happy numbers between 1 and 100 are : {}".format(ans))
```

<b>Q5. </b>Write a Python program to determine whether the given number is a Harshad Number?

```python
num = input("Enter a number : ")
res = 0
for i in num:
    res = res + int(i)
if int(num)%res == 0:
    print("{} is a Harshad Number".format(num))
else:
    print("{} is not a Harshad Number".format(num))
```
