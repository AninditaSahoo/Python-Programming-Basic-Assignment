<b>Q2. </b>Write a Python program to do arithmetical operations addition and division.?

<b>Q3. </b>Write a Python program to find the area of a triangle?



```python
b = int(input("Breadth : "))
h = int(input("Height : "))
print("as breadth = {} and height = {} hence area = {}".format(b,h,(0.5*b*h)))
```

<b>Q4. </b>Write a Python program to swap two variables?

```python
a,b = 4,5
print("Before swapping")
print("a = {}, b = {}".format(a,b))
a,b = b,a
print("After swapping")
print("a = {}, b = {}".format(a,b))
```
