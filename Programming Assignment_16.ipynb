<b><u> Question1</u></b>

Write a function that stutters a word as if someone is struggling to read it. The first two letters are repeated twice with an ellipsis ... and space after each, and then the word is pronounced with a question mark ?.

<b>Examples</b>

stutter("incredible") ➞ "in... in... incredible?"

stutter("enthusiastic") ➞ "en... en... enthusiastic?"

stutter("outstanding") ➞ "ou... ou... outstanding?"

<b>Hint :-</b> Assume all input is in lower case and at least two characters long.

```python
def stutter(word):
    return word[0:2]+"... "+word[0:2]+"... "+word+"?"
word = input("Enter the word : ")
print(stutter(word))
```

<b>Answer: 
<br>Enter the word : enthusiastic
<br>en... en... enthusiastic?</b>

<b><u>Question 2</u></b>

Create a function that takes an angle in radians and returns the corresponding angle in degrees rounded to one decimal place.

<b>Examples</b>

radians_to_degrees(1) ➞ 57.3

radians_to_degrees(20) ➞ 1145.9

radians_to_degrees(50) ➞ 2864.8

```python
import math
def radians_to_degrees(value):
    return round(value*(180/math.pi),1)
value = float(input("Enter the Temperature in degree : "))
print(radians_to_degrees(value))
```

<b>Answer: 
<br>Enter the Temperature in degree : 50
<br>2864.8</b>

<b><u> Question 3 </u></b>

In this challenge, establish if a given integer num is a Curzon number. If 1 plus 2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon number.

Given a non-negative integer num, implement a function that returns <b>True</b> if num is a Curzon number, or <b>False</b> otherwise.

<b>Examples:</b>

<table style="border:1px solid black;" align="left">
    <tr><th> is_curzon(5) ➞ True  </th></tr>
    <tr><td> # 2 ** 5 + 1 = 33  </td></tr>
    <tr><td> # 2 * 5 + 1 = 11</td></tr>
    <tr><td> # 33 is a multiple of 11 </td></tr>
</table>

<table style="border:1px solid black;" align="left">
    <tr><th> is_curzon(10) ➞ False  </th></tr>
    <tr><td> # 2 ** 10 + 1 = 1025  </td></tr>
    <tr><td> # 2 * 10 + 1 = 21</td></tr>
    <tr><td> # 1025 is not a multiple of 21 </td></tr>
</table>

<table style="border:1px solid black;" align="left">
    <tr><th> is_curzon(14) ➞ True  </th></tr>
    <tr><td> # 2 ** 14 + 1 = 16385  </td></tr>
    <tr><td> # 2 * 14 + 1 = 29</td></tr>
    <tr><td> # 16385 is a multiple of 29 </td></tr>
</table>

```python
def is_curzon(num):
    if (2**num+1)%(2*num+1) == 0:
        return True
    else:
        return False
num = int(input("Enter a number : "))
print(is_curzon(num))
```

<b>Answer : 
<br>Enter a number : 14
<br>True</b>

<b><u>Question 4 </u></b>

Given the side length x find the area of a hexagon.


<b>Examples:</b>

area_of_hexagon(1) ➞ 2.6

area_of_hexagon(2) ➞ 10.4

area_of_hexagon(3) ➞ 23.4

```python
import math
def area_of_hexagon(num):
    return round((3*math.sqrt(3)*(num**2))/2,1)
num = int(input("Enter a side length : "))
print(area_of_hexagon(num))
```

<b>Answer: 
<br>Enter a side length : 2
<br>10.4</b>

<b><u>Question 5</u></b>

Create a function that returns a base-2 (binary) representation of a base-10 (decimal) string number. To convert is simple: ((2) means base-2 and (10) means base-10) 010101001(2) = 1 + 8 + 32 + 128.

Going from right to left, the value of the most right bit is 1, now from that every bit to the left will be x2 the value, value of an 8 bit binary numbers are (256, 128, 64, 32, 16, 8, 4, 2, 1).

<b>Examples:</b>
<table style="border:1px solid black;" align="left">
    <tr><th> binary(1) ➞ "1"  </th></tr>
    <tr><td> # 1*1 = 1  </td></tr>
</table>

<table style="border:1px solid black;" align="left">
    <tr><th> binary(5) ➞ "101"  </th></tr>
    <tr><td> # 1*1 + 1*4 = 5 </td></tr>
</table>

<table style="border:1px solid black;" align="left">
    <tr><th> binary(10) ➞ "1010"</th></tr>
    <tr><td> # 1*2 + 1*8 = 10  </td></tr>
</table>

```python
def binary(num):
    return str(bin(num)).split("b")[1]
num = int(input("Enter a number : "))
print(binary(num))
```

<b>Answer : 
<br>Enter a number : 5
<br>101</b>
